```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pyfonts import load_font

font_regular = load_font(
    font_url="https://github.com/SavaMaksimtsev/raphinha/raw/main/GoogleSans-Regular.ttf"
)
font_bold = load_font(
    font_url="https://github.com/SavaMaksimtsev/raphinha/raw/main/GoogleSans-Bold.ttf"
)

data = {
    "": [
        "", "", "", " ", "",
        " ", " ", "", " ", "",
        "", "", "", "", "",
        "", "", "", "", " "
    ],
    "2023-24": [
        23, 41, 37, 39, 0, 45, 26, 43, 37, 41,
        35, 26, 36, 33, 29, 0, 0, 22, 25, 26
    ],
    "2024-25": [
        21, 20, 19, 19, 18, 17, 16, 16, 14, 14,
        12, 12, 10, 10, 10, 10, 9, 8, 7, 4
    ]
}

df = pd.DataFrame(data)

df_sorted = df.sort_values(by="2024-25", ascending=True)

teams = df_sorted[""]
values_2023_24 = df_sorted["2023-24"]
values_2024_25 = df_sorted["2024-25"]

y = np.arange(len(teams))

plt.figure(figsize=(10, 8))

plt.barh(y - 0.2, values_2023_24, height=0.4, color='#67139D', label='2023-24')
plt.barh(y + 0.2, values_2024_25, height=0.4, color='#4EA075', label='2024-25')

plt.yticks(y, teams, font=font_regular)
plt.xlabel('')
plt.title('-      -2024/25  2023/24', font=font_bold, size=16)
plt.legend()
plt.gca().spines['top'].set_color('white')
plt.gca().spines['right'].set_color('white')

plt.tight_layout()
plt.savefig("injuries.png")
plt.show()
```
